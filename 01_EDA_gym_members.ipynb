```python
# import numpy as np
# import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns

df = pd.read_csv("gym_members_exercise_tracking.csv")
df.head()
```

```python
# Shape of dataset
print("Rows & Columns:", df.shape)

# Column data types
print("\nData Types:")
print(df.dtypes)

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Basic statistics
df.describe()
```

```python
# Distribution of key columns
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle("Distribution of Key Variables", fontsize=16, fontweight='bold')

sns.histplot(df['Age'], ax=axes[0,0], color='steelblue', kde=True)
axes[0,0].set_title("Age Distribution")

sns.histplot(df['Calories_Burned'], ax=axes[0,1], color='tomato', kde=True)
axes[0,1].set_title("Calories Burned")

sns.histplot(df['BMI'], ax=axes[0,2], color='mediumseagreen', kde=True)
axes[0,2].set_title("BMI Distribution")

sns.histplot(df['Session_Duration (hours)'], ax=axes[1,0], color='mediumpurple', kde=True)
axes[1,0].set_title("Session Duration")

sns.countplot(data=df, x='Workout_Type', ax=axes[1,1], palette='Set2')
axes[1,1].set_title("Workout Type Count")

sns.countplot(data=df, x='Gender', ax=axes[1,2], palette='Set1')
axes[1,2].set_title("Gender Distribution")

plt.tight_layout()
plt.show()
```

```python
# Average calories burned per workout type
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Workout_Type', y='Calories_Burned', palette='Set2', errorbar=None)
plt.title("Average Calories Burned by Workout Type", fontsize=14, fontweight='bold')
plt.xlabel("Workout Type")
plt.ylabel("Avg Calories Burned")
plt.tight_layout()
plt.savefig("calories_by_workout.png", dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Map experience levels to labels
df['Experience_Label'] = df['Experience_Level'].map({1: 'Beginner', 2: 'Intermediate', 3: 'Advanced'})

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Experience Level Analysis", fontsize=14, fontweight='bold')

# Calories by experience
sns.barplot(data=df, x='Experience_Label', y='Calories_Burned', 
            palette='coolwarm', errorbar=None, ax=axes[0],
            order=['Beginner', 'Intermediate', 'Advanced'])
axes[0].set_title("Avg Calories Burned by Experience Level")
axes[0].set_xlabel("Experience Level")
axes[0].set_ylabel("Avg Calories Burned")

# Session duration by experience
sns.barplot(data=df, x='Experience_Label', y='Session_Duration (hours)', 
            palette='coolwarm', errorbar=None, ax=axes[1],
            order=['Beginner', 'Intermediate', 'Advanced'])
axes[1].set_title("Avg Session Duration by Experience Level")
axes[1].set_xlabel("Experience Level")
axes[1].set_ylabel("Avg Session Duration (hours)")

plt.tight_layout()
plt.savefig("experience_analysis.png", dpi=150, bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Body Composition Analysis", fontsize=14, fontweight='bold')

# Fat percentage by gender
sns.boxplot(data=df, x='Gender', y='Fat_Percentage', 
            palette='Set1', ax=axes[0])
axes[0].set_title("Fat Percentage by Gender")
axes[0].set_xlabel("Gender")
axes[0].set_ylabel("Fat Percentage (%)")

# BMI vs Fat Percentage scatter
sns.scatterplot(data=df, x='BMI', y='Fat_Percentage', 
                hue='Gender', palette='Set1', alpha=0.5, ax=axes[1])
axes[1].set_title("BMI vs Fat Percentage")
axes[1].set_xlabel("BMI")
axes[1].set_ylabel("Fat Percentage (%)")

plt.tight_layout()
plt.savefig("body_composition.png", dpi=150, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
numeric_df = df.select_dtypes(include=np.number)
correlation = numeric_df.corr()

sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm', 
            linewidths=0.5, annot_kws={'size': 9})
plt.title("Correlation Heatmap", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig("correlation_heatmap.png", dpi=150, bbox_inches='tight')
plt.show()
```
